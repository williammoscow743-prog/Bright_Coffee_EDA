# TASK 02 — FEATURE ENGINEERING

```python
# Revenue column
df["revenue"] = df["unit_price"] * df["transaction_qty"]

# Hour column
df["hour"] = pd.to_datetime(
    df["transaction_time"],
    format="%H:%M:%S"
).dt.hour

# Month column
df["month"] = df["transaction_date"].dt.month

df.head()
```

```python
plt.figure(figsize=(9,6))

sns.histplot(df["unit_price"], kde=True)

plt.title("Distribution of Unit Price")
plt.xlabel("Unit Price")
plt.ylabel("Frequency")

plt.show()
```

### Distribution Observation
The unit price distribution appears right-skewed, meaning most products are lower-priced

# 2. Revenue by Store Location

```python
store_revenue = df.groupby("store_location")["revenue"].sum()

plt.figure(figsize=(10,6))

store_revenue.plot(kind="bar")

plt.title("Total Revenue by Store Location")
plt.xlabel("Store Location")
plt.ylabel("Revenue")

plt.show()
```

```python
sns.boxplot(data=df, x="store_location", y="Revenue")
plt.show()
```

# 3. Transactions by Hour

```python
hourly_transactions = df.groupby("hour")["transaction_id"].count()

plt.figure(figsize=(10,6))

hourly_transactions.plot(kind="line", marker="o")

plt.title("Transactions by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Transactions")

plt.grid(True)

plt.show()
```

### Business Hours Observation
The coffee shop is busiest during the morning and lunchtime hours, 
which likely reflects commuter and lunch customer behaviour.

# 1. Matplotlib Version

```python
import plotly.express as px
```

```python
cat_rev = df.groupby("product_category")["Revenue"].sum().reset_index()

cat_rev.columns = ["category","revenue"]

cat_rev = cat_rev.sort_values("revenue",ascending = False)

cat_rev
```

```python
fg = px.bar(cat_rev, x="category", y="revenue",
            color="revenue",
            # color_continuous_scale="Greens"
            )
fg.show()
```

```python
monthly = df.groupby(["Month","store_location"])["Revenue"].sum().reset_index()
monthly
```

```python
fig2 = px.line(monthly, x="Month", y="Revenue",
               color = "store_location",
               markers=True)
fig2.show()
```

```python
#Monthly revenue trend
monthly_trend = df.groupby("Month")["Revenue"].sum()
monthly_trend
```

## 2. Seaborn Version

```python
cat_rev = df.groupby("product_category")["revenue"].sum().reset_index()
```

```python
plt.figure(figsize=(10,6))

plt.bar(
    cat_rev["product_category"],
    cat_rev["revenue"]
)

plt.title("Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Revenue")

plt.xticks(rotation=45)

plt.show()
```

### Visualisation Library Comparison

Matplotlib required the most code because many chart formatting options must be manually configured. 
Seaborn produced the best default appearance with cleaner styling and less code. 
Plotly provided interactive features such as hover labels and zooming, which are useful for dashboards 
and presentations. In real-world business applications, Plotly would be ideal for interactive reporting 
tools where users need to explore the data themselves.
